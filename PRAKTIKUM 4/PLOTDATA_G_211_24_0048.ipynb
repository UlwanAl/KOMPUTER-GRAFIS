```python
import matplotlib.pyplot as plt
import numpy as np

# Set axis limits and grid
plt.axis([0, 140, 0, 100])
plt.axis('on')
plt.grid(True)

# Data
t = [20, 40, 60, 80, 100, 120]               # time
T = [30, 35, 43, 55, 70, 85]                 # temperature
p = [2, 3.4, 5.3, 7.3, 9.6]                  # pressure
v = [0.6, 0.58, 0.54, 0.46, 0.35, 0.2]       # volume

# Scale pressure by 10 and volume by 100 for plotting
pp = [pi * 10 for pi in p]
vv = [vi * 100 for vi in v]

# Plot data
plt.plot(t, T, color='r', label='Temperature', marker='o')       # Temperature
plt.plot(t[:5], pp, color='b', label='Pressure', marker='s')     # Scaled pressure (same length as p)
plt.plot(t, vv, color='g', label='Volume', marker='d')           # Scaled volume

# Add legend
plt.legend(loc='upper left')

# Annotate pressure scale on right (manually)
for y in np.arange(0, 100+1, 20):
    a = y / 10
    a = str(a)
    plt.text(142, y, a, color='b')  # position text on far right

# Label axes
plt.xlabel('Time (hrs)')
plt.ylabel('Temperature °K', color='r')
plt.text(151, 65, 'Pressure (psi)', rotation=90, color='b')

# Annotate volume scale (manually)
for y in np.arange(100, -1, -20):
    a = y / 100
    a = str(a)
    plt.text(162, y, a, color='g')
    plt.text(159, y + 2, '_', color='g')

# Add minor ticks for volume
for y in np.arange(1, 99, 3):
    plt.text(157, y, '_', color='g')

# Volume label
plt.text(170, 65, 'Volume (cm³)', rotation=90, color='g')

# Title
plt.title('Compression Test Results')

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Contoh data (ganti dengan data asli jika perlu)
t = np.array([0, 1, 2, 3, 4, 5])
v = np.array([0, 2, 4.1, 6.1, 7.9, 10.2])

# ===== Regresi Linier =====
n = len(v)

c1 = np.sum(t) / n  # rata-rata t
c2 = np.sum(v) / n  # rata-rata v
a = np.multiply(v, t)
c3 = np.sum(a)
a = np.multiply(t, t)
c4 = np.sum(a)

A = (c3 - n * c1 * c2) / (c4 - n * c1 * c1)
B = c2 - A * c1

# Atur batas sumbu x dan y
plt.axis([0, 150, 0, 100])
plt.axis('on')  # (opsional, ini tidak diperlukan karena default-nya 'on')
plt.grid(True)  # Aktifkan grid


# ===== Plot hasil regresi =====
for tp in np.arange(t[0], t[5], 2):
    vp = A * tp + B
    vp = vp * 100  # penskalaan (bisa dihilangkan jika tidak diperlukan)
    plt.scatter(tp, vp, color='g', s=1)

# --------- definisikan fungsi y1 terhadap x ---------
x = np.arange(0, 150, 1)
y1 = 10 + np.exp(0.035 * x)

# ===== Hitung RMS error =====
sumee = 0
for i in range(len(t)):
    e = v[i] - (A * t[i] + B)
    ee = e * e
    sumee += ee
rms = np.sqrt(sumee / n)

# ===== Tambah label ke grafik =====
plt.text(60, 28, 'v=At+B', color='g')
plt.arrow(78, 30, 6, 6, head_length=3, head_width=1.5, color='g', linewidth=0.5)

# Data
t = [20, 40, 60, 80, 100, 120]               # time
T = [30, 35, 43, 55, 70, 85]                 # temperature
p = [2, 3.4, 5.3, 7.3, 9.6]                  # pressure
v = [0.6, 0.58, 0.54, 0.46, 0.35, 0.2]       # volume

# Scale pressure by 10 and volume by 100 for plotting
pp = [pi * 10 for pi in p]
vv = [vi * 100 for vi in v]

# Plot data
plt.plot(t, T, color='r', label='Temperature', marker='o')       # Temperature
plt.plot(t[:5], pp, color='b', label='Pressure', marker='s')     # Scaled pressure (same length as p)
plt.plot(t, vv, color='g', label='Volume', marker='d')           # Scaled volume

# Add legend
plt.legend(loc='upper left')

# Annotate pressure scale on right (manually)
for y in np.arange(0, 100+1, 20):
    a = y / 10
    a = str(a)
    plt.text(142, y, a, color='b')  # position text on far right

# Label axes
plt.xlabel('Time (hrs)')
plt.ylabel('Temperature °K', color='r')
plt.text(151, 65, 'Pressure (psi)', rotation=90, color='b')

# Annotate volume scale (manually)
for y in np.arange(100, -1, -20):
    a = y / 100
    a = str(a)
    plt.text(162, y, a, color='g')
    plt.text(159, y + 2, '_', color='g')

# Add minor ticks for volume
for y in np.arange(1, 99, 3):
    plt.text(157, y, '_', color='g')

# Volume label
plt.text(170, 65, 'Volume (cm³)', rotation=90, color='g')

# Titik awal dan akhir garis regresi
vp1 = A * t[0] + B
vp1 = '%7.4f' % vp1
plt.text(2, 64, vp1, color='g')

vp2 = A * t[5] + B
vp2 = '%7.4f' % vp2
plt.text(122, 25, vp2, color='g')

# Tampilkan nilai A dan B
Ap = '%7.5f' % A
plt.text(65, 18, 'A=', color='g')
plt.text(72, 18, Ap, color='g')

Bp = '%7.5f' % B
plt.text(65, 12, 'B=', color='g')
plt.text(73, 12, Bp, color='g')

# Tampilkan nilai RMS error
rms_str = '%7.3f' % rms
plt.text(95, 3, 'RMS error=', color='g')
plt.text(123, 3, rms_str, color='g')

plt.xlabel('t')
plt.ylabel('v')
plt.title('Regresi Linier v vs t')
plt.grid(True)
plt.show()
```
